```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
losses = np.load("losses.npy")
```

```python
average_losses = [np.mean(epoch) for epoch in losses]
```

```python
plt.title("Average Cross-Entropy Loss Per Epoch (40 Guitar Hero 5 songs)")
plt.xlabel("Epoch")
plt.ylabel("Average Cross-Entropy Loss")
plt.ylim(0,1.5)
plt.plot(list(range(len(average_losses))), average_losses)
plt.show()
```

```python
losses2 = np.load("losses2.npy")
average_losses2 = [np.mean(epoch) for epoch in losses2]
```

```python
plt.title("Average Cross-Entropy Loss Per Epoch (One Guitar Hero 5 song)")
plt.xlabel("Epoch")
plt.ylabel("Average Cross-Entropy Loss")
plt.ylim(0,1.5)
plt.plot(list(range(len(average_losses2))), average_losses2)
plt.show()
```

```python
total_average_losses_40 = np.concatenate((average_losses3, average_losses5))
plt.title("Average Loss Per Epoch (40 Guitar Hero 5 songs, 300+ Epochs)")
plt.xlabel("Epoch")
plt.ylabel("Average Loss")
plt.ylim(0,1.5)
plt.plot(list(range(len(total_average_losses_40[:150]))), total_average_losses_40)

x = list(range(len(total_average_losses_40[:150])))
y = total_average_losses_40[:150]
coefficients = np.polyfit(x, y, 1)
plt.plot(x, np.poly1d(coefficients)(x), label=f'Line of Best Fit (Slope={coefficients[0]:.4f})')
plt.legend()

plt.show()
```
