```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
losses = np.load("losses.npy")
```

```python
average_losses = [np.mean(epoch) for epoch in losses]
```

```python
plt.title("Average Cross-Entropy Loss Per Epoch (40 Guitar Hero 5 songs)")
plt.xlabel("Epoch")
plt.ylabel("Average Cross-Entropy Loss")
plt.ylim(0,1.5)
plt.plot(list(range(len(average_losses))), average_losses)
plt.show()
```
